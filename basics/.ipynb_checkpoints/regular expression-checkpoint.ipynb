```python
# Simply searching for the exact matching substring
input_str = 'Hi! this is Amar. My contact is 908-209-8998 (Amar)'
'908-209-8998' in input_str
```

```python
# Regular Expression is needed when we don't know exactly what we are searching for. We just know the format or pattern of our query.
# for eg. ###-###-####
import re
```

```python
# Simple string search
query_str = 'Amar'

result = re.search(query_str, input_str)

# Span of the matched query_str
print(result.span()) # it returns the last matched string
```

```python
# Get all available matches
result = re.findall(query_str, input_str)
result
```

```python
# Find the span of all matches
for result in re.finditer(query_str, input_str):
    print(result.span())
```

### Pattern Matching

```python
input_string = "In its discrete form, a hidden Markov process can be visualized as a generalization of the urn problem with replacamarement (where each item from the urn is returned to the original urn before the next step).[6] Consider this example: in a room that is not visible to an observer there is a genie. The room contains urns X1, X2, X3, ... each of which contains a known mix of balls, each ball labeled y1, y2, y3, ... . The genie chooses an urn in that room and randomly draws a ball from that urn. It then puts the ball onto a conveyor belt, where the observer can observe the sequence of the balls but not the sequence of urns from which they were drawn. The genie    has some procedure to choose urns; the choice of the urn for Amar the n-th ball depends only upon a random number and the Amarchoice of the urn for the (n  1)-th ball. Contact 923-098-1233, 2230912301"
```

```python
# Find a mobile number - use groups in pattern
patter = r'(\d{3})-(\d{3})-(\d{4})'
sresult = re.search(patter, input_string)
sresult.groups()
```

```python
# Get groups
sresult = re.search(patter, input_string)
sresult.groups()
```

```python
# Use pipe operator
pattern = r'(\d+|\W{2})' # one or more digit OR non numeric maximum 2 character

for result in re.finditer(pattern, input_string):
    print(result.span(), result)
```

```python
# Find words containing substrings
pattern = r'(\w*|\s)(amar|Amar)(\w*|\s)' # find the string containing Amar as a substring

for result in re.finditer(pattern, input_string, ):
    print(result.span(), result)
```

```python
# Use wild card character
pattern = r'\w*[^\W*]' # keep only valid words | exclude non alphanumeric

for result in re.finditer(pattern, input_string):
    print(result.span(), result)
```
